```python
# Task 1: User Login Check

# stored credentials
correct_username = "admin"
correct_password = "1234"

# input from user
username = input("Enter username: ")
password = input("Enter password: ")

# validation
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# output results
print("Total Students:", len(marks))
print("Passed:", pass_count)
print("Failed:", fail_count)
```

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

#final clean list
print("Cleaned Names:", cleaned_names)
```

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("This message is longer than 10 characters")
```

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Total ERROR messages:", error_count)
```
